```python
# Importing PANDAS library
import pandas as pd

# Specifying the path to the main .csv archive
csv_path = r"C:\Users\Jose Manuel\Downloads\nacional.csv"

# reading csv as a dataframe
df = pd.read_csv(csv_path, sep=';', on_bad_lines='skip')

# showing the first 5 rows
df.head()
```

```python
# Specifying the path to the second .xlsx archive to make the merge
xls_path = r"C:\Users\Jose Manuel\Downloads\centros.xlsx"

# reading xlsx as a dataframe
df_centro = pd.read_excel(xls_path)

# showing the first 5 rows
df_centro.head()
```

```python
# Specifying the path to the second .xlsx archive to make the merge
centros_csv_path = r"C:\Users\Jose Manuel\Downloads\centros.csv"

# reading csv as a dataframe
df_centro_csv = pd.read_csv(centros_csv_path)

# showing the first 5 rows
df_centro_csv.head()
```

```python
# Merging df y df_centro_csv by cod_centro column
df_merge_1 = pd.merge(df, df_centro_csv, on="cod_centro", how="left")
df_merge_1.head()
```

```python
# Merging df_merge_1 and df_centro bythe columns centro_nuevo & CODIGO
df_merge_2 = pd.merge(df_merge_1, df_centro, left_on="centro_nuevo", right_on="CODIGO", how="left")
df_merge_2.head()
```

```python
# Selecting columns to keep
df_final = df_merge_2[["nacionalidad", "cedula", "primer_apellido", "segundo_apellido", "primer_nombre", "segundo_nombre", "cod_centro", "MUNICIPIO", "ESTADO"]]
df_final.head()
```
